```python
import pandas as pd

german = pd.read_csv('/content/drive/MyDrive/transactions_germany.csv')
german.head()
```

```python
# Import required modules from mlxtend
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
# Display frequent items sets
frequent_itemsets = apriori(german, min_support = 0.04, use_colnames = True)
# Increase column width
pd.set_option('display.max_colwidth', 100)
# Display the frequent itemsets
frequent_itemsets
```

```python
# Add a new column storing the length of each itemset
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets
```

```python
# Show frequent itemsets with length 2 and have support of at least 4%
frequent_itemsets[(frequent_itemsets['length'] == 2) & \
 (frequent_itemsets['support'] >= 0.04)]
```

```python
# Generate association rules. 
rules = association_rules(frequent_itemsets, metric = "confidence", min_threshold = 0.5)
rules
```

```python
# Generate association rules. 
rules = association_rules(frequent_itemsets, metric = "lift", min_threshold = 1)
rules
```
